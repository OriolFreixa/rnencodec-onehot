<b>Build your own sequence</b>

```python
#Used to spread a parameter vector over time - generates with unchanging parameter values.
def repeat_values(values, T, device=None, dtype=torch.float32):
    x = torch.tensor(values, dtype=dtype, device=device)
    return x.unsqueeze(0).expand(T, -1)

# creates the one-hot vector of classes
def one_hot(length, pos): # pos in [0, length-1]
    if not (0 <= pos < length):
        raise IndexError("pos out of range")
    return [1 if i == pos else 0 for i in range(length)]

```

```python
import soundfile as sf
# ['ChirpPattern', 'DSApplause', 'DSBugs', 'DSPeepers', 'DSPistons', 'DSWind', 'TokWotalDuet']
rootdir="synth4kad"

start = 0.0  #pval
end = 1.0    #pval (inclusive) 
step = 0.05   #step
N = int(round((end - start) / step)) + 1
xs = np.linspace(start, end, N)

nunclasses = 7 
numtimepoints = offline_render_frames # 2 for testing
numrepeats=5

numhops=int(offline_render_frames/g_hopsize)

#counter = 0
start = time.perf_counter()

for classnum in range(0,nunclasses) :
    clslist=one_hot(7,classnum) 
    
    classdir = os.path.join(rootdir, f"g{classnum}")
    os.makedirs(classdir, exist_ok=True)
    
    for pval in xs:
        paramlist=[pval, *clslist]
        #warmup_paramlist=[.5, *clslist] # for warmup, extreme values can lead to silence generation 
        byo_param_seq = repeat_values(paramlist, numtimepoints)
        for r in range(0,numrepeats+1) :


            bazrnn = RNNGeneratorSoft.from_checkpoint(checkpoint_path, model_config, data_config, enc_model, g_chunksize, g_hopsize, sample_mode_outside=g_sample_mode, top_k_outside=g_top_n, temperature_outside =g_temperature)
            #warm up prevents noise bursts at the begininng of the audio run
            # _ = bazrnn.warmup(warmup_paramlist, 100, sigma=1) #xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx

            byo_trialsegments=[]

            for i in range(numhops):
                params_seq = byo_param_seq[i*g_hopsize : i*g_hopsize + g_hopsize] # p.view(1, -1).expand(g_hopsize, -1)  # shape (T, 2), view with stride 0
                #print(f' for hop {i}, params_seq is {params_seq}')
                a=bazrnn.getNextAudioHop(params_seq)  # returns a 1D float32 array like in your example
                byo_trialsegments.append(a)
            byo_audio = np.concatenate(byo_trialsegments, dtype=np.float32)  # one allocation
            


            
            filename = f"class_{classnum}_v{pval:.2f}_r{r:.2f}.wav"
            filepath = os.path.join(classdir, filename)
          
            print(f'-------  file {filepath}') 
            sf.write(filepath, byo_audio[-keep_duration * sr :], sr)

        
        #counter = counter + 1
elapsed = time.perf_counter() - start
print(f'Total time to render all files  {elapsed}')                       
```
